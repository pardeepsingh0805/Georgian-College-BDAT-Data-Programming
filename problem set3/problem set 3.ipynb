```python
#step 1
import pandas as pd
import numpy as np
import urllib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## step 2
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

#step 3
users = pd.read_csv(url, delimiter='|')
print(users)
```

```python
## step 4

age = users[['occupation','age']].groupby("occupation").mean()
print(age)
```

```python
## step 5

df = users[['occupation','gender']]
males = df.loc[df['gender']== 'M']
occupation = df.groupby('occupation').count()
male_count = males.groupby("occupation").count()
ratio = male_count/occupation
ratio.sort_values(ascending= False, by='gender')
```

```python
## step 6

max_age = users[['occupation','age']].groupby("occupation").max()
print("################ MAX AGE ############")
print(max_age)

min_age = users[['occupation','age']].groupby("occupation").min()
print("\n################ MIN AGE ############")
print(min_age)
```

```python
## step 7
df = users[['occupation','gender','age']].groupby(['occupation','gender']).mean('age')
df
```

```python
## step 8
df = users[['occupation','gender','age']]
percentage = df.groupby('occupation')['gender'].value_counts(normalize=True)*100
percentage
```

```python
# step 1
import pandas as pd



# step 2
data = pd.read_csv('wind.txt', delimiter='\s+')

# step 3
data.rename(columns={"Yr":"Year", "Mo":"Month", "Dy":"Day"}, inplace=True)

data
```

```python
# step 4
def Year():
    data['Year'] += 1900 #

Year() # Fix year by converting it in 4-digit value

data
```

```python
## step 5

data.index = pd.to_datetime(pd.concat([data['Year'], data['Month'], data['Day']], axis=1))
data.drop(['Year','Month','Day'], axis=1, inplace=True)
data
```

```python
## step 6 for each
data.isna().sum()
```

```python
# entire record
data.isnull().sum().sum()
```

```python
## step 7 computing missing values
data.notnull().sum().sum()
```

```python
## step 8

data.mean().mean()
```

```python
## step 9

loc_stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 
loc_stats.rename(columns={0:'Min', 1:'Max', 2:'Mean', 3:'Std'}, inplace=True) 

loc_stats
```

```python
## step 10

days = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)
days.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)
days
```

```python
# step 11

wind_jan = data[data.index.month == 1].mean()
wind_jan
```

```python
# step 13

monthly = data.resample('M').ffill()

monthly
```

```python
## step 14 
weekly = data.resample('W').ffill()

weekly
```

```python
## step 1
import pandas as pd

# step 2 and 3

chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', delimiter='\t')
chipo.head(10)
```

```python
# step5: number of observations or rows
print(f"The number of rows are : {chipo.shape[0]}")
```

```python
# step6: number of columns
print(f"The number of columns are : {chipo.shape[1]}")
```

```python
# step 7

for col in chipo.columns:
    print(col)
```

```python
# step 8

chipo.index
```

```python
## step 9

most_ordered_item = (chipo.groupby('item_name').sum()).sort_values(['quantity'], ascending=False).head(1)

most_ordered_item
```

```python
## step 10

most_ordered_item[['quantity']]
```

```python
# step 11
desc_item = chipo.groupby(['choice_description']).sum().sort_values('quantity', ascending=False).head(1)

desc_item
```

```python
## step 12
print(f'Number of items ordered in total = {chipo.quantity.sum()}')
```

```python
## STEP 13

print(f"the previous data type was: {chipo.item_price.dtype}")
## conversion
chipo.item_price = chipo.item_price.apply(lambda x : float(x[1:-1]))

# new dtype
print(f"the new data type was: {chipo.item_price.dtype}")
```

```python
## step 14

revenue = (chipo.quantity * chipo.item_price).sum()

print(f'Total revenue is ${revenue}')
```

```python
## step 15

orders = chipo.order_id.value_counts().count()
print(f'Total number of orders are {orders}')
```

```python
## step 16

temp_chipo = chipo.loc[:,('order_id', 'quantity', 'item_price')]

temp_chipo['revenue'] = temp_chipo['quantity'] * temp_chipo['item_price']

avg_amt = temp_chipo.groupby(by=['order_id']).sum().mean()['revenue']

print(f'The average amount per order is {avg_amt}')
```

```python
## step 17

items_sold = chipo.item_name.value_counts().count()

print(f'Total different items sold = {items_sold}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data
```

```python
fig = plt.figure()
data.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])
plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014', color = 'k')
plt.xlabel('Year', color = 'k', fontsize = 17)
plt.ylabel('Capita', color = 'k', fontsize = 17)

plt.show()
```

<img src="./q7.png" width="500"/>

```python
filters = (data['Year'] == 1900) | (data['Year'] == 1950) | (data['Year'] == 2000) 

df = data.loc[filters, ['Year','Marriages_per_1000','Divorces_per_1000']]

df[{'Marriages_per_1000','Divorces_per_1000'}].plot.bar()

plt.title("Marriages and Divorces per capita in the U.S. between 1900, 1950 and 2000", color = 'k')

plt.xlabel('Year', color = 'k', fontsize = 17)
plt.ylabel('Capita', color = 'k', fontsize = 17)

plt.xticks((0, 1, 2), ("1900", "1950", "2000"), rotation = 0)
```
